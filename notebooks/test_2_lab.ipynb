```python
df = get_table_test(data)
def clean_dataframe(df):
    # Identify rows where the entire row except the first column is empty
    is_colspan_row = df.iloc[:, 1:].isna().all(axis=1) | (df.iloc[:, 1:] == '').all(axis=1)
    
    # For rows identified as 'colspan' rows, fill the first column downwards
    df.loc[is_colspan_row, 'Показатель'] = df.loc[is_colspan_row, 'Показатель'].ffill()
    
    # Drop the 'colspan' rows that only contain the first column data
    df = df[~is_colspan_row]
    
    return df
cleaned_df = clean_dataframe(df)
cleaned_df
```
