```python
import os

def count_files_in_folder(folder_path):
    # Initialize file count
    file_count = 0
    
    # Walk through the directory
    for root, dirs, files in os.walk(folder_path):
        file_count += len(files)
    
    return file_count



```

```python
import os

def rename_files_in_folder(folder_path, base_name='name'):
    # List all files in the directory
    files = os.listdir(folder_path)
    
    # Sort files to maintain consistent renaming order
    # files.sort()
    
    # Initialize a counter
    i = 1
    
    for filename in files:
        # Construct the full file path
        old_file_path = os.path.join(folder_path, filename)
        
        # Check if it's a file (not a directory)
        if os.path.isfile(old_file_path):
            # Extract file extension
            file_extension = os.path.splitext(filename)[1]
            
            # Construct the new file name
            new_filename = f"{base_name}{i}{file_extension}"
            new_file_path = os.path.join(folder_path, new_filename)
            
            # Rename the file
            os.rename(old_file_path, new_file_path)
            
            print(f'Renamed: {filename} -> {new_filename}')
            
            # Increment the counter
            i += 1

# Example usage
folder_path = '/Users/maximzabelin/My/edc/hse4/data/med_records/t_fol'  # Replace with your folder path
rename_files_in_folder(folder_path, base_name='file_')
```

```python
import os

def rename_files_in_folder(folder_path, base_name='name', log_file='renaming_log.txt'):
    # List all files in the directory
    files = os.listdir(folder_path)
    
    # Sort files to maintain consistent renaming order
    # files.sort()
    
    # Initialize a counter
    i = 1
    
    # Open the log file in write mode
    with open(log_file, 'w') as log:
        for filename in files:
            # Construct the full file path
            old_file_path = os.path.join(folder_path, filename)
            
            # Check if it's a file (not a directory)
            if os.path.isfile(old_file_path):
                # Extract file extension
                file_extension = os.path.splitext(filename)[1]
                
                # Construct the new file name
                new_filename = f"{base_name}{i}{file_extension}"
                new_file_path = os.path.join(folder_path, new_filename)
                
                # Rename the file
                os.rename(old_file_path, new_file_path)
                
                # Log the renaming information
                log_message = f'{filename} : {new_filename}\n'
                print(f'{filename} : {new_filename}')
                log.write(log_message)
                
                # Increment the counter
                i += 1


```

```python

folder_path = '/Users/maximzabelin/My/edc/hse4/data/med_records/DocsNew1'  
rename_files_in_folder(folder_path, base_name='file1_', log_file='renaming_log_DocsNew1.txt')

folder_path = '/Users/maximzabelin/My/edc/hse4/data/med_records/DocsNew2'  
rename_files_in_folder(folder_path, base_name='file2_', log_file='renaming_log_DocsNew2.txt')

folder_path = '/Users/maximzabelin/My/edc/hse4/data/med_records/DocsNew3'  
rename_files_in_folder(folder_path, base_name='file3_', log_file='renaming_log_DocsNew3.txt')
```
